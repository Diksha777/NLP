```python
def missing_values_analysis(df):
    na_columns_ = [col for col in df.columns if df[col].isnull().sum() > 0]
    n_miss = df[na_columns_].isnull().sum().sort_values(ascending = True)
    ratio_ = (df[na_columns_].isnull().sum() /df.shape[0]* 100).sort_values(ascending = True)
    missing_df = pd.concat([n_miss, np.round(ratio_, 2)], axis =1, keys=['Missing Values','Ratio'])
    missing_df = pd.DataFrame(missing_df)
    return missing_df

def check_dataframe(df, head=5, tail=5):
    print("SHAPE".center(82,'~'))
    print('Rows: {}'.format(df.shape[0]))
    print('columns: {}'.format(df.shape[1]))
    print("TYPES".center(82,'~'))
    print(df.dtypes)
    print("".center(82,'~'))
    print(missing_values_analysis(df))
    print("DUPLICATE VALUES".center(83,'~')) 
    print(df.duplicated().sum())
    print("QUANTILES".center(82,'~')) 
    print(df.quantile([0, 0.05, 0.50, 0.95, 0.90, 1]).T)
    
check_dataframe(df)
    
    
    
        
    
```

```python
def check_class(dataframe):
    nunique_df = pd.DataFrame({'Variable': dataframe.columns,
                             'Classes': [dataframe[i].nunique() \
                                        for i in dataframe.columns]})
    nunique_df = nunique_df.sort_values('Classes', ascending = False)
    nunique_df = nunique_df.reset_index(drop= True)
    return nunique_df

check_class(df)
    
```

```python
constraints = ['#B34D22', '#EBE00C','#1FEB0C','#0C92EB','#EB0CD5']
def categorical_variable_summary(df, column_name):
    fig = make_subplots(rows =1, col = 2,
                        subplot_titles=('Countplot', 'Percentage'),
                        specs =[[{"types": "xy"},{'types':'domain'}]])
    
    fig.add_trace(go.Bar( y = df[column_name].value_counts().values.tolist(),
                          x = [str(i) for i in df[column_name].value_counts().index],
                          text = df[column_name].value_counts().values.tolist(),
                          textfont = dict(size=14),name = column_name,
                          textposition = 'auto',
                        showlegend = False,
                          marker = dict(color = constraints,
                                        line= dict(color='#DBE6EC',
                                                   width = 1))),
                  row = 1, col = 1)
    fig.add_trace(go.Pie(labels = df[column_name].value_counts().keys(),
                           values = df[column_name].value_counts().values,
                         textfont = dict(size = 18),
                         textposition = 'auto',
                         showlegend = False,name = column_name,
                          marker = dict(colors=constraints)),
                  row = 1, col = 2)
    
    fig.update_layout(title={'text': column_name,
                             'y':0.9,
                             'x':0.5,
                            'xanchor':'center',
                            'yanchor':'top'},
                      template ='plotly_white')
    inplot(fig)
                            
    
                         
                        
```

```python
def categorical_variable_summary(data, column):
    """
    This function takes a DataFrame and a column name as inputs,
    and returns a summary of the categorical variable in the column.

    Parameters:
        data (pandas.DataFrame): The DataFrame to analyze
        column (str): The name of the column to summarize

    Returns:
        A pandas DataFrame with the following columns:
        - category: The unique categories in the column
        - count: The number of observations in each category
        - percent: The percentage of observations in each category
    """
    # Group the data by the column and count the number of observations in each group
    group_counts = data.groupby(column).size().reset_index(name='count')

    # Calculate the percentage of observations in each category
    group_counts['percent'] = group_counts['count'] / len(data) * 100

    # Sort the categories by count in descending order
    group_counts = group_counts.sort_values('count', ascending=False)

    return group_counts
```
